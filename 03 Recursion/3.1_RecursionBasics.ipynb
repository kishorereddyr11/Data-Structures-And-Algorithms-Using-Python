```python
## Function Calling Itself
def fun():
    print("Hello World")
    fun()
fun()
```

```python
def fun(n):
    if n==0: return
    print("Hello World!")
    fun(n-1)
fun(4)
```

```python
##Factorial Using Recursion
def fact(n):
    if n==1:
        return 1
    return n*fact(n-1)
fact(5)
```

```python
## Print numbers n to 1
def Print(n):
    if n==0:
        return 
    print(n)
    Print(n-1)
Print(5)
```

```python
## Print numbers 1 to n
def Print(n):
    if n==0:
        return
    Print(n-1)
    print(n)
Print(5)
```

```python
## Print Sum from 1 to n
def Sum(n):
    if n==1:
        return 1
    return n+Sum(n-1)
Sum(5)
```

```python
## a Raised To the Power b
def pow(a,b):
    if b==1:
        return a
    return a*pow(a,b-1)
pow(2,3)
```

```python
## Nth Fibonacci Number
def fibo(n):
    if n==1 or n==2:
        return 1
    return fibo(n-1) + fibo(n-2)
```

```python
## Printing Fibonacci Sequence
for i in range(1,11):
    print(fibo(i))
```

```python
## Power Function Logerthmic
def pow(a,b):
    if b == 1:
        return a
    ans = pow(a,b//2)
    if b % 2 == 0:
        return ans*ans
    if b % 2 != 0:
        return ans*ans*2
pow(2,3)
```

```python
## Starir Path
def Stair(n):
    if n == 1:
        return 1 # if there is one stair one one step he can climb
    if n == 2:
        return 2 # if there is Two Stairs there are two ways to climb
    return Stair(n-1)+Stair(n-2)
Stair(4)
```

```python
## Maize Path
def maze(sr,sc,er,ec):
    if sr > er or sc > ec : 
        return 0
    if sr == er and sc == ec :
        return 1
    RightWays =  maze(sr,sc+1,er,ec)
    DownWays =  maze(sr+1,sc,er,ec)
    return RightWays + DownWays
print(maze(0,0,2,2))

```

```python
## Print Maze Path
def maze(sr,sc,er,ec,str):
    if sr > er or sc > ec :
        return 0
    if sr == sc and sc == ec :
        print(str)
        return 1
    return maze(sr,sc+1,er,ec,str+"R ") + maze(sr+1,sc,er,ec,str+"D ")
maze(0,0,2,2,"")
```

```python
## Pre in Post 
## Predict Output
def pip(n):
    if n==0:
        return
    print("Pre :",n)
    pip(n-1)
    print("In ",n)
    pip(n-1)
    print("Post ",n)
pip(3)
```

```python
## Tower of Hanoi
def Hanoi(n,a,b,c):
    if n==0:
        return
    Hanoi(n-1,a,c,b)
    print(f"{a} to {c}")
    Hanoi(n-1,b,a,c)
Hanoi(3,1,2,3)
```

```python
## Traversing in array using Recursion
def traverse(lst,idx):
    if idx == len(lst):
        return
    print(lst[idx],end=" ")
    traverse(lst,idx+1)
traverse([1,2,3,4,5],0)
```

```python
## Finding Maximum in array using Recursion
def Max(lst,idx):
    if idx==len(lst):
        return float('-inf')
    return max(lst[idx],Max(lst,idx+1))
Max([2,5,8,4,3],0)
```
